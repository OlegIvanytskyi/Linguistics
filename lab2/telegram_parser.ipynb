```python
import datetime
from pytz import timezone
from telethon.sync import TelegramClient
import pandas as pd
```

```python
timezone = timezone("Europe/Kiev")
birthday = datetime.datetime(2019, 11, 29, tzinfo=timezone)
```

```python
bad_kpi = []
good_kpi = []
```

```python
async with TelegramClient('anon', '1156046', 'bba70b2ba116afabda6f930db7b33a00') as client:
    async for message in client.iter_messages('kpi_666'):
        message.date = message.date.replace(tzinfo=timezone)
        if len(bad_kpi) >= 100 and message.date < birthday:
            break
        if message.message and len(message.message.split()) < 2:
            continue
        if message.message:
            bad_kpi.append(message)
    async for message in client.iter_messages('kpi_777'):
        message.date = message.date.replace(tzinfo=timezone)
        if len(good_kpi) >= 100 and message.date < birthday:
            break
        if message.message and len(message.message.split()) < 2:
            continue
        if message.message:
            good_kpi.append(message)
```

```python
print(len(bad_kpi))
print(len(good_kpi))
```

```python
messages_bad = []
dates_bad = []
labels_bad = []

i = 0
for post in bad_kpi[::-1]:
    if len(post.message.split()) < 2:
        continue
    messages_bad.append(post.message)
    dates_bad.append(post.date)
    labels_bad.append(-1)
    i += 1
    if i == 100:
        break
print(len(messages_bad))
```

```python
messages_good = []
dates_good = []
labels_good = []

i = 0
for post in good_kpi[::-1]:
    if len(post.message.split()) < 2:
        continue
    messages_good.append(post.message)
    dates_good.append(post.date)
    labels_good.append(1)
    i += 1
    if i == 100:
        break
print(len(messages_good))
```

```python
channel = []
messages = []
dates = []
labels = []
for i in range(100):
    channel.append('bad kpi')
    channel.append('good kpi')
    messages.append(messages_bad[i].lower())
    messages.append(messages_good[i].lower())
    dates.append(dates_bad[i])
    dates.append(dates_good[i])
    labels.append(labels_bad[i])
    labels.append(labels_good[i])
print(len(messages))
```

```python
data = {'channel': channel, 'text': messages, 'datetime': dates, 'label': labels}
```

```python
df = pd.DataFrame(data=data)
df.index += 1
```
